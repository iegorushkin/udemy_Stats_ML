```python
# import libraries
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
## simulate data
rng = np.random.default_rng()
N = 66

# generate correlated data
x = rng.standard_normal(size=N)
y = x + rng.standard_normal(size=N)
#
```

```python
## plot the data
plt.plot(x, y, "kp", markerfacecolor="b", markersize=12)
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
## compute covariance

# precompute the means
meanX = np.mean(x)
meanY = np.mean(y)

# the loop method
covar1 = 0
for i in range(N):
    covar1 = covar1 + (x[i]-meanX)*(y[i]-meanY)
# and now for the normalization
covar1 = covar1/(N-1)

# the linear algebra method
xCent = x - meanX
yCent = y - meanY
covar2 = np.dot(xCent, yCent) / (N-1)

# the Python method
covar3 = np.cov(np.vstack((x, y)))
# в матрице, передаваемой np.cov, строки - переменные,
# столбцы - наблюдения (тут - 2 переменные, 66 наблюдений)
print(f"{covar1}, \n{covar2}, \n{covar3}")
```

```python
## now for correlation

# the long method
corr_num = np.sum((x - meanX) * (y - meanY))
corr_den = np.sum((x-meanX)**2) * np.sum((y - meanY)**2)
corr1 = corr_num/np.sqrt(corr_den)

# the Python method
corr2 = np.corrcoef(np.vstack((x, y)))  # автокорреляция по одной из диагоналей 
# corr2 = np.corrcoef(x, y)  # то же самое 

print(f"{corr1}, \n{corr2}")
```

```python
## correlation as normalized covariance

xn = stats.zscore(x, ddof=1)
yn = stats.zscore(y, ddof=1)
corr3 = np.dot(xn, yn) / (N-1)

print(corr3)
```

```python
## 2D t-value space based on r (Pearson correlation) and n (number of samples)

# define parameters
r = np.linspace(-1, 1, 250)
n = np.round(np.linspace(5, 200, 100))

# initialize t- and p-value matrices
tmatrix = np.zeros((len(r), len(n)))
pmatrix = np.zeros((len(r), len(n)))

for ri in range(len(r)):
    for ni in range(len(n)):
        # the t-value, split into num/den
        num = r[ri]*np.sqrt(n[ni] - 2)
        den = 1 - r[ri]**2
        tmatrix[ri, ni] = num / den
        pmatrix[ri, ni] = 1 - stats.t.cdf(np.abs(num / den), n[ni]-2)
        
# question: Why the warning message?
# Потому что в формуле для расчёта t-матрицы при r=1 получается ноль в знаменателе
```

```python
# Soooo curious to see it!
plt.imshow(tmatrix, vmin=-3, vmax=3, extent=[n[0], n[-1], r[0], r[-1]],
           aspect='auto', origin='lower')
plt.colorbar()
# pmatrix < 0.05 задаёт матрицу значений по оси Z, в которой True = 1, False = 0;
# 1 - по идее, количество уровней контурных линий. Не уверен, насколько оно тут надо.
plt.contour(pmatrix < 0.05, 1, colors='k', extent=[n[0], n[-1], r[0], r[-1]])
plt.xlabel('Sample size')
plt.ylabel('Correlation coefficient')
plt.title('T-values from different r-n combos')
plt.show()
```

```python
# final note on statistical significance
# Можно воспользоваться встроенной в numpy функцией для расчёта коэффициента корреляции 
# и соответствующего p-значения
r, p = stats.pearsonr(x, y)
print(r, p)
```
