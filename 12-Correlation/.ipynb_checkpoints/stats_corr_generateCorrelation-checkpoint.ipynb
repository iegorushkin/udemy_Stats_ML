```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## simulate data

# random generator
rng = np.random.default_rng()

# data simulation parameters
N = 100  # number of samples
r = 0.4   # desired correlation coefficient

# start with random numbers
x = rng.standard_normal(size=N)
y = rng.standard_normal(size=N)

# impose the correlation on y
y = x*r + y*np.sqrt(1 - r**2)

# plot the data
plt.plot(x, y, 'kp', markerfacecolor='b', markersize=12)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
## compute the empirical correlation

empR = np.corrcoef(x, y)  # массив 2х2

print(f'Desired r={r}, empirical r={empR[0, 1]}')
```

```python
## Test the errors (разница между желаемым и эмпирическим коэфф-ми коррелции)
# as a function of N

# range of sample sizes
Ns = np.round(np.linspace(5, 400, 123)).astype(int)
# theoretical correlation coefficient (fixed)
r = 0.6
# initialize
corrs = np.zeros(len(Ns))

# run the experiment!
for ni in range(len(Ns)):
    x = rng.standard_normal(size=Ns[ni])
    y = x*r + rng.standard_normal(size=Ns[ni])*np.sqrt(1 - r**2)
    corrs[ni] = (r - np.corrcoef(x, y)[0, 1])**2  # square error

plt.stem(Ns, corrs, 'ko-', basefmt=" ")
plt.xlabel('Sample size')
plt.ylabel('Squared divergence')
plt.show()
```
