```python
# import libraries
import matplotlib.pyplot as plt
import numpy.matlib as matlib
import numpy as np
```

```python
## simulate data

# simulation parameters
N = 1000  # time points
M = 20  # channels

# time vector (radian units)
t = np.linspace(0, 6*np.pi, N)

# relationship across channels (imposing covariance)
chanrel = np.sin(np.linspace(0, 2*np.pi, M))

# create the data
rng = np.random.default_rng()
data = np.zeros(shape=(M, N))
for mi in range(M):
    data[mi:] = np.sin(t) * chanrel[mi]
data = data + rng.standard_normal(size=(M, N))

# Разница между data[mi:] и data[mi, :]
# 1) data[mi:] вернет все строки начиная с строки mi включительно
# 2) data[mi, :] вернет строку с индексом mi

# Таким образом, каждый "канал" после первого домножается на несколько chanrel[mi]
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(8, 8))

# Выбираем цветовую карту (например, 'gist_rainbow')
cmap = plt.get_cmap('gist_rainbow')

for i in range(M):
    color = cmap(i / M)  # Выбираем цвет из цветовой карты
    ax1.plot(t, data[i, :] + i*4, color=color)
ax1.set(xlim=[0, t[-1]], xlabel="Time (a.u.)", ylabel="Channel")
# Задаем собственные метки на осях
ax1.set_yticks(np.linspace(0, M*4, 5))  # Задаем позиции меток на оси X
ax1.set_yticklabels(['1', '5', '10', '15', '20'])  # Задаем текст меток

plot=ax2.imshow(data, cmap=cmap, aspect='auto',
                vmin=-2, vmax=2, extent=[t[0], t[-1], 0, M], origin='lower')
ax2.set(xlabel='Time (a.u.)', ylabel='Channel')
ax2.set_yticks(np.linspace(0, M, 5))  # Задаем позиции меток на оси X
ax2.set_yticklabels(['1', '5', '10', '15', '20'])  # Задаем текст меток
plt.show()

# We see 2 distinct groups of channels -
# 1 through 10, 11 through 20
```

```python
## now compute the covariance matrix

# note the size of the output!
dataCovMat = np.cov(data.T, rowvar=False)
# If `rowvar` is True (default), then each row represents a
# variable, with observations in the columns. Otherwise, the relationship
# is transposed: each column represents a variable, while the rows
#  contain observations.
plt.imshow(dataCovMat, vmin=-.5, vmax=.5)
plt.title('Data covariance matrix')
plt.xlabel('??')
plt.ylabel('??')
plt.show()
```

```python
## and now the correlation matrix

# note the size of the output!
dataCorrMat = np.corrcoef(data)

plt.imshow(dataCorrMat, vmin=-.5, vmax=.5)
plt.title('Data correlation matrix')
plt.xlabel('??')
plt.ylabel('??')
plt.show()
```
