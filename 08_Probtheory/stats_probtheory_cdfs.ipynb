```python
# import libraries
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
## example using log-normal distribution

# variable to evaluate the functions on
x = np.linspace(0, 5, 1001)

# note the function call pattern...
p1 = stats.lognorm.pdf(x, 1)
c1 = stats.lognorm.cdf(x, 1)

p2 = stats.lognorm.pdf(x, 0.1)
c2 = stats.lognorm.cdf(x, 0.1)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

# draw the pdfs
# ax[0].plot(x, p1/np.sum(p1)) # question: why divide by sum here?
# ax[0].plot(x, p2/np.sum(p2))

# To transform pdf into pmf, p1 is scaled by np.sum(p1) 
# It is some sort of computational hack, this way we make sure that cumsum goes to 1
# Another way is multiplying by (x[1] - x[0])
# np.sum(p1)*(x[1] - x[0]) â‰ˆ 1;

ax[0].plot(x, p1*np.mean(np.diff(x)))
ax[0].plot(x, p2*np.mean(np.diff(x)))
ax[0].set_ylabel('probability')
ax[0].set_title('pdf(x)')

# draw the cdfs
ax[1].plot(x, c1)
ax[1].plot(x, c2)
ax[1].set(xlim=[0, 5], ylabel='cum. probability', xlabel='x-value',
          title='cdf(x)')
plt.show()
```

```python
## computing the cdf from the pdf

# compute the cdf
c1x = np.cumsum(p1*np.mean(np.diff(x)))

fig, ax = plt.subplots()
ax.plot(x, c1, label='analytical')
ax.plot(x, c1x, '--', label='empirical')
ax.set(xlim=[0, 5], ylabel='cum. probability', xlabel='x-value',
          title='cdf(x)')
ax.legend()
plt.show()
```
