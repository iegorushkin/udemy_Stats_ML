```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## compute empirical probability function

# continous signal (technically discrete!)
N = 10004
rng = np.random.default_rng()
# Brownian noise
datats1 = np.cumsum(np.sign(rng.standard_normal(size=N)))
datats2 = np.cumsum(np.sign(rng.standard_normal(size=N)))

# let's see what they look like
plt.plot(np.arange(N), datats1, linewidth=2)
plt.plot(np.arange(N), datats2, linewidth=2)
plt.xlabel("N")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# discretize using histograms
nbins = 50

y, x = np.histogram(datats1, nbins)
x1 = (x[1:] + x[:-1])/2
y1 = y/np.sum(y)

y, x = np.histogram(datats2, nbins)
x2 = (x[1:] + x[:-1])/2
y2 = y/np.sum(y)

plt.plot(x1, y1, x2, y2, linewidth=3)
plt.legend(('ts1', 'ts2'))
plt.xlabel('Data value')
plt.ylabel('Probability')
plt.grid()
plt.show()
```
