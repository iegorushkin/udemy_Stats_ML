```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## create some data

N = 40
rng = np.random.default_rng()

data = rng.standard_normal(size=N)
# force some outliers
data[data < -2] = -data[data < -2]**2
data[data > 2]  =  data[data > 2]**2

# also need the mean-centered data
dataMC = data - np.mean(data)

# and plot them (not it ;) )
fig, ax = plt.subplots()
ax.plot(data, 'k^', markerfacecolor='y', markersize=12)
# ax.set_xticks([])
# ax.set_yticks([])
ax.set_xlabel('Data index')
ax.set_ylabel('Data value')
ax.grid()
plt.show()
```

```python
np.percentile(np.abs(dataMC), 100 - 5)
```

```python
## option 1: remove k% of the data

# percent of "extreme" data values to remove
trimPct = 5 # in percent

# identify the cut-off (note the abs() )
datacutoff = np.percentile(np.abs(dataMC), 100 - trimPct)
print(f"{100 - trimPct} percentile = {datacutoff}")

# find the exceedance data value indices 
data2cut = np.nonzero(np.abs(dataMC) > datacutoff)[0]
print(f"Indices of outliers: {data2cut}")

# and mark those off
ax.plot(data2cut, data[data2cut], 'rx', markersize=15)
fig
```

```python
## Numpy SQUEEZE example
# Создаем массив с избыточной размерностью
a = np.array([[[1], [2], [3]]])

print("Исходный массив:")
print(a)
print("Форма исходного массива:", a.shape)

# Используем squeeze для удаления избыточной размерности
b = np.squeeze(a)

print("\nМассив после использования squeeze:")
print(b)
print("Форма массива после использования squeeze:", b.shape)
```

```python
## option 2: remove k most extreme values

# number of "extreme" data values to remove
k2remove = 3  # in number

# find the exceedance data values
# -1 меняет порядок элементов массива; 
# т.е. в datasortIdx хранятся индексы значений от большего к меньшему
datasortIdx = np.argsort(np.abs(dataMC), axis=0)[::-1] 

# data2cut = np.squeeze(datasortIdx[:k2remove]) # а зачем тут squeeze?
print(datasortIdx[:k2remove])
print(np.squeeze(datasortIdx[:k2remove]))

# Ликвидируем первые k-индесков 
data2cut = (datasortIdx[:k2remove])
```

```python
# and mark those off
ax.plot(data2cut, data[data2cut], 'go', markersize=15, alpha=.5)

# finally, add a legend
ax.legend(('All data', f'{100-trimPct}%% threshold', 'k2remove-value threshold'),
          framealpha=1, loc='lower right', labelspacing=0.8)
fig
```
