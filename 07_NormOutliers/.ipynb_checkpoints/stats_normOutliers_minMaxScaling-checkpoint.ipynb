```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create some data
N = 42
rng = np.random.default_rng()
data = np.log(rng.uniform(size=N))*234 + 934
```

```python
# get min and max
dataMin = np.min(data)
dataMax = np.max(data)

# now min-max scale
dataS = (data - dataMin) / (dataMax - dataMin)

# now plot
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
fig.subplots_adjust(wspace=0.3)

# такой сложный х в plot позволяет избежать перекрытия данных
# и лучше почувствовать их распределение
temp = rng.standard_normal(size=N)/20
ax[0].plot(1 + temp, data, 'ks')
ax[0].set_xlim([0, 2])
ax[0].set_xticks([])
ax[0].set_ylabel('Original data scale')
ax[0].set_title('Original data')

ax[1].plot(1 + temp, dataS, 'ks')
ax[1].set_xlim([0, 2])
ax[1].set_xticks([])
ax[1].set_ylabel('Unity-normed data scale')
ax[1].set_title('Scaled data')

plt.show()
```

```python
## show that scaling doesn't affect the relative values

plt.plot(data, dataS, 'ks')
plt.xlabel('Original')
plt.ylabel('Scaled')
plt.grid()
plt.show()
```

```python
## any abitrary data range

# step 1 is to [0, 1] normalize as above

# step 2:
newMin = 4
newMax = 8.7

dataSS = newMin + dataS*(newMax - newMin) 

# test it!
print([np.min(dataSS), np.max(dataSS)])
```
