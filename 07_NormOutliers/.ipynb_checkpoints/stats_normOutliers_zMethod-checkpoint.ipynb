```python
import numpy as np
import scipy.stats as stats
from statsmodels import robust
import matplotlib.pyplot as plt
```

```python
## create some data

N = 70
rng = np.random.default_rng()
data = rng.standard_normal(N)
# ко всем значениям меньше -1 добавляем +2
data[data < -1] = data[data < -1] + 2
# все значения больше двух возводим в квадрат - создаем аномальные значения
data[data > 2] = data[data > 2]**2; # try to force a few outliers
data = data*200 + 50  # change the scale for comparison with z (?)
# Вообще, можно было бы использовать нормальное распределение с заданными m и std 

# convert to z
dataZ = (data - np.mean(data)) / np.std(data)

#### specify the z-score threshold
zscorethresh = 3
```

```python
# plot the data
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

ax[0].plot(data, 'k^', markerfacecolor='w', markersize=12)
ax[0].set_xticks([])
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Orig. scale')

# then plot the zscores
ax[1].plot(dataZ, 'k^', markerfacecolor='w', markersize=12)
ax[1].plot([0, N], [zscorethresh, zscorethresh], 'r--')
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Z distance')
plt.show()
```

```python
## identify outliers

# find 'em! (indeces)
outliers = np.nonzero(np.abs(dataZ) > zscorethresh)[0]
# Possible way, but not preferred:
# outliers = np.where(np.abs(dataZ) > zscorethresh)[0]  

# and cross those out
ax[0].plot(outliers, data[outliers], 'x', color='r', markersize=20)
ax[1].plot(outliers, dataZ[outliers], 'x', color='r', markersize=20)
fig # instead of plt.show()
```

```python
## iterative method

# pick a lenient threshold just for illustration
zscorethresh = 2
dataZ = (data - np.mean(data)) / np.std(data)

colorz = 'brkmgy'
numiters = 0 # iteration counter

while True:
    # На первой итерации считается z-scorе от z-score
    # datamean = 0, datastd = 1
    
    # Дальше странновато, конечно - вместо пересчёта z-score от data
    # считается z-score от z-score без учёта определенных ранее аутлаеров 
    
    # convert to z
    datamean = np.nanmean(dataZ)
    datastd  = np.nanstd(dataZ)
    dataZ = (dataZ - datamean) / datastd
    
    # find data values to remove (outliers)
    toremove = np.abs(dataZ) > zscorethresh
    
    # break out of while loop if no points to remove
    if np.sum(toremove) == 0:
        break
    # otherwise, mark the outliers in the plot
    else:
        plt.plot(np.nonzero(toremove)[0], dataZ[toremove], linestyle='', marker='s',
                 markersize=12, color=colorz[numiters])
        # plt.plot(np.nonzero(toremove)[0], dataZ[toremove], '%sx'%colorz[numiters], markersize=12)
        # "Ликвидируем" выброс в данных (но не удаляем)
        dataZ[toremove] = np.nan
    
    # replot updated dataset
    plt.plot(dataZ, 'k^', markersize=12, markerfacecolor=colorz[numiters],
             label=f'iteration {numiters}')
    numiters = numiters + 1

plt.xticks([])
plt.ylabel('Z-score')
plt.xlabel('Data index')
plt.legend()
plt.show()

#### the data points to be removed
# ищем индексы nan-значений (это те ребята, которые были аутлаерами)
removeFromOriginal = np.nonzero(np.isnan(dataZ))[0]
print(removeFromOriginal)
```

```python
## modified Z for non-normal distributions

# pick a lenient threshold just for illustration
zscorethresh = 2

# compute modified z
dataMed = np.median(data)
dataMAD_statsmodels = robust.mad(data, c=1)
# если не менять с (которое по умолчанию равно scipy.stats.norm.ppf(3/4.)=0.6745),
# то dataMz нужно считать без stats.norm.ppf(.75)
dataMAD_scipy = stats.median_abs_deviation(data)
print(f"dataMAD_statsmodels: {dataMAD_statsmodels},",
      f"dataMAD_scipy: {dataMAD_scipy}")

# Percent point function - ppf; 
# stats.norm.ppf(.75) - значение 3 квартиля стандартного нормально распр.
dataMz = stats.norm.ppf(.75)*(data-dataMed) / dataMAD_scipy

# plot the data
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

ax[0].plot(data, 'k^', markerfacecolor='w', markersize=12)
ax[0].set_xticks([])
#ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Orig. scale')

# then plot the zscores
ax[1].plot(dataMz, 'k^', markerfacecolor='w', markersize=12)
ax[1].plot([0, N], [zscorethresh, zscorethresh], 'r--')
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Median dev. units (Mz)')
plt.show()
```
