```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# create data for the bar plot

# data sizes
m = 30  # rows
n = 6  # columns
# random generator
rng = np.random.default_rng()

# generate data
data = np.zeros((m, n))

for i in range(n):
    data [:, i] = 30 * rng.standard_normal(size=m) * \
        (2 * i / (n - 1) - 1) ** 2 + (i + 1) ** 2

# for i in range(n):
#     data[:, i] = 30 * np.random.randn(m) * \
#         (2 * i / (n - 1) - 1) ** 2 + (i + 1) ** 2
```

```python
# show the bars!

fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# 'naked' bars
# тут и далее усредняются значения в столбцах таблицы data
ax[0].bar(range(n), np.mean(data, axis=0))
ax[0].set_title("Bar plot")

# just the error bars
ax[1].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1),
               marker="s", linestyle="",)
ax[1].set_title("Errorbar plot")

# both
ax[2].bar(range(n), np.mean(data, axis=0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1),
               marker=".", linestyle="", color="k",)
ax[2].set_title("Error + bar plot")
plt.show()
```

```python
# manually specify x-axis coordinates; make custom ticks for our bars

xcrossings = [1, 2, 4, 5, 6, 9]

fig, ax = plt.subplots()
ax.bar(xcrossings, np.mean(data, axis=0))
ax.errorbar(xcrossings, np.mean(data, axis=0), np.std(data, axis=0, ddof=1),
            marker=".", linestyle="", color="k",)
ax.set(title="Bar + error plot", xticks=xcrossings);
```

```python
# note about bars from matrices

# data are groups (rows) x property (columns)
m = [[2, 5, 4, 3], [1, 1, 8, 6]]
# using pandas dataframe
df = pd.DataFrame(m, columns=["prop 0", "prop 1", "prop 2", "prop 3"])

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

# conceptualizing the data as <row> groups of <columns>
ax[0, 0].imshow(m)
df.plot(ax=ax[1, 0], kind="bar")
ax[1, 0].set_title("Grouping by rows")

# now other orientation (property x group)
ax[0, 1].imshow(np.array(m).T)
df.T.plot(ax=ax[1, 1], kind="bar")
ax[1, 1].set_title("Grouping by columns")

plt.show()
```
