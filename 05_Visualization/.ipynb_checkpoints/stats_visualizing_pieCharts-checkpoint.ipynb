```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data for the plot

nbins = 5
totalN = 100

# set of values between 1 and 5
rawdata = np.ceil(np.logspace(np.log10(1 / 2), np.log10(nbins - 0.01), totalN))
print(f"rawdata: \n{rawdata}")
```

```python
# prepare data for pie chart
uniquenums = np.unique(rawdata)  # vector of unique values in raw data
print(f"uniquenums: {uniquenums}")
data4pie = np.zeros(len(uniquenums))
print(f"data4pies: {data4pie}")
```

```python
# how many times each unique value is encountered in rawdata
for i in range(len(uniquenums)):
    data4pie[i] = np.sum(rawdata == uniquenums[i])
print(f"data4pies: {data4pie}")
```

```python
# show the pie chart
# label is a percent of whole dataset
plt.pie(data4pie, labels=(100 * data4pie / totalN))
plt.show()
```

```python
# another option
plt.pie(data4pie, labels=["zero", "one", "two", "three", "four"],
        explode=[0, 0.1, 0, 0.15, 0],)
# explode : array-like, default: None
#     If not *None*, is a ``len(x)`` array which specifies the fraction
#     of the radius with which to offset each wedge.
plt.show()
```

```python
# for continuous data

# generate log-normal distribution
data = np.exp(np.random.randn(1000) / 10)

# generate bins using histogram
histout = np.histogram(data, bins=6)

# and show that as a pie chart
plt.pie(histout[0])
plt.show()
```
