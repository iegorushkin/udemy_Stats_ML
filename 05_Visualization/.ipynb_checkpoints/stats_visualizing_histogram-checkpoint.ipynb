```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the histogram

# number of data points
n = 1000

# generate data - log-normal distribution
rng = np.random.default_rng()
data = np.exp(rng.standard_normal(n) / 2)
# data = np.exp(np.random.randn(n) / 2)
```

```python
# show as a histogram

# number of histogram bins
k = 40

plt.hist(data, bins=k)
plt.show()
```

```python
# another option
y, x = np.histogram(data, bins=k)
print(len(y), type(y))
print(len(x), type(x))
# x is bigger than y by 1
# y contains heights of bars, x conteins their left and right borders!

# bin centers
xx = (x[1:] + x[:-1]) / 2
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
fig.subplots_adjust(hspace=0.3)

# If the data has already been binned and counted, use `~.bar` or
# `~.stairs` to plot the distribution::
#     counts, bins = np.histogram(x)
#     plt.stairs(counts, bins)
# Alternatively, plot pre-computed bins and counts using ``hist()`` by
# treating each bin as a single point with a weight equal to its count::
#     plt.hist(bins[:-1], bins, weights=counts)
ax[0].hist(x[:-1], x, weights=y)
ax[0].set_title("Histogram inffered from np.histogram")

# histogram as a line
ax[1].plot(xx, y)
ax[1].set_title("Line plot from histogram data");
```
