```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the plot

# number of data points
n = 1000

# generate log-normal distribution
rng = np.random.default_rng()
data1 = np.exp(rng.standard_normal(n) / 2)
data2 = np.exp(rng.standard_normal(n) / 10)
data3 = np.exp(rng.standard_normal(n) / 2 + 1)
```

```python
## plots of their histograms

# number of histogram bins
k = 20

plt.hist(data1, bins=k)
plt.hist(data2, bins=k)
plt.hist(data3, bins=k)

plt.show()
```

```python
plt.style.available
```

```python
# histogram discretization for the datasets
y1, x1 = np.histogram(data1, bins=k)
xx1 = (x1[:-1] + x1[1:]) / 2  # middle of a bin
y1 = y1 / np.sum(y1)  # convert to probability (proportion)

y2, x2 = np.histogram(data2, bins=k)
xx2 = (x2[:-1] + x2[1:]) / 2
y2 = y2 / np.sum(y2)

y3, x3 = np.histogram(data3, bins=k)
xx3 = (x3[:-1] + x3[1:]) / 2
y3 = y3 / np.sum(y3)

with plt.style.context("seaborn-v0_8-whitegrid"):
    plt.plot(xx1, y1, "s-", label="data1")
    plt.plot(xx2, y2, "o-", label="data2")
    plt.plot(xx3, y3, "^-", label="data3")
    plt.xlabel("Value")
    plt.ylabel("Probability")
    plt.legend()
    plt.show()
```
