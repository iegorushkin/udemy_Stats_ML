```python
# import libraries
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
# simulate data
rng = np.random.default_rng()

popN = int(1e7)  # lots and LOTS of data!!

# the data (note: non-normal!)
population = (4*rng.standard_normal(size=popN))**2

# we can calculate the exact population mean
popMean = np.mean(population)
```

```python
# let's see it
fig, ax = plt.subplots(2, 1, figsize=(12, 8))

# only plot every 1000th sample
ax[0].plot(population[::1000], 'k.')
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Data value')

ax[1].hist(population, bins='fd')
ax[1].set_xlabel('Data value')
ax[1].set_ylabel('Count')
ax[1].set_ylim([0, 300000])
plt.show()
```

```python
# draw a random sample

# parameters
samplesize = 40
confidence = 95  # in percent

# samplesize случайных int из [0, popN)
# indices of our sample
irandSamples = rng.integers(low=0, high=popN, size=samplesize)
# sample itself - random values taken from the population
randSamples = population[irandSamples]
# sample mean
samplemean = np.mean(randSamples)
# sample std
samplestd = np.std(randSamples, ddof=1)

# compute confidence intervals
citmp = (1 - confidence/100) / 2
confint = samplemean + \
    stats.t.ppf([citmp, 1-citmp], samplesize-1)*samplestd/np.sqrt(samplesize)
# confint - массив, содержащий t-значения для левой и правой границы доверительного интервала
# По сути, формулка для расчёта citmp - это некоторое переусложнение.
# Если мы хотим получить значения t-распределения, соответствующие 97.5% и 2.5%
# (тогда длина CI будет 95%), можно просто сразу написать их
# в stats.t.ppf([0.025, 0.975], samplesize-1)
```

```python
# graph everything
fig, ax = plt.subplots(figsize=(10, 6))

# Рисуем зеленый прямоугольник
y = np.array([[confint[0], 0], [confint[1], 0],
             [confint[1], 1], [confint[0], 1]])
# print(y)  # out of curiosity
p = Polygon(y, facecolor='g', alpha=0.3)
ax.add_patch(p)

# now add the lines
ax.plot([popMean, popMean], [0, 1.5], 'k:', linewidth=3)
ax.plot([samplemean, samplemean], [0, 1], 'r--', linewidth=3)
ax.set_xlim([popMean-30, popMean+30])
ax.set_xlabel('Data values')
ax.set_yticks([])
ax.set_ylim([0, 1.5])
ax.legend((f'{confidence}% CI region', f'True mean ({np.round(popMean, 2)})',
           f'Sample mean ({np.round(samplemean, 2)})'))
plt.show()
```

```python
# repeat for large number of samples

# parameters
samplesize = 50
confidence = 95  # in percent
numExperiments = 5000
withinCI = np.zeros(numExperiments)  # Сюда записываем результаты экспериментов

# part of the CI computation can be done outside the loop
citmp = np.round((100 - confidence)/200, 3)
#                    2.5%  97.5%
CI_T = stats.t.ppf([citmp, 1-citmp], samplesize-1)
# print(CI_T)
# same as using plus-minus stats.t.ppf(1-citmp, samplesize-1)
sqrtN = np.sqrt(samplesize)

for expi in range(numExperiments):
    # compute sample mean and CI as above
    # indices of our sample
    irandSamples = rng.integers(low=0, high=popN, size=samplesize)
    # sample itself - random values taken from the population
    randSamples = population[irandSamples]
    # sample mean
    samplemean = np.mean(randSamples)
    # sample std
    samplestd = np.std(randSamples, ddof=1)
    # array of CI limits
    confint = samplemean + CI_T*samplestd/sqrtN

    # determine whether the True mean is inside this CI
    if popMean > confint[0] and popMean < confint[1]:
        withinCI[expi] = 1

print(f'{100*np.mean(withinCI)}% of sample C.I.' +
      's contained the true population mean.')
```
