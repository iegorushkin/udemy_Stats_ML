```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np

## make plots interactive
# import ipympl
# matplotlib ipympl
# %matplotlib widget
```

```python
## create some data distributions

# the distributions
N = 10001   # number of data points
nbins = 50  # number of histogram bins

rng = np.random.default_rng()
d1 = rng.standard_normal(N) - 1
d2 = 3*rng.standard_normal(N)
d3 = rng.standard_normal(N) + 1

# need their histograms
# reminder - this done to plot a line through the centers of bins
y1, x1 = np.histogram(d1, nbins)
x1 = (x1[1:] + x1[:-1]) / 2

y2, x2 = np.histogram(d2, nbins)
x2 = (x2[1:] + x2[:-1]) / 2

y3, x3 = np.histogram(d3, nbins)
x3 = (x3[1:] + x3[:-1])/ 2

# plot them
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
plt.plot(x3, y3, 'k')
plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.grid()
plt.show()
```

```python
## side note:

# the distribution's parameters
meanval = 10.2
stdval = 7.5
numsamp = 10000
# random generator with a fixed seed (for repeatability)
rng_test = np.random.default_rng(seed=10)

# this
data1 = rng_test.normal(loc=meanval, scale=stdval, size=numsamp)
# is equivalent to
data2 = rng_test.standard_normal(size=numsamp)*stdval + meanval
# let's check this

bins = 60
hist1, bin_edges1 = np.histogram(data1, bins)
bin_edges1 = (bin_edges1[1:] + bin_edges1[:-1]) / 2
hist2, bin_edges2 = np.histogram(data2, bins)
bin_edges2 = (bin_edges2[1:] + bin_edges2[:-1]) / 2

# plot
fig, ax = plt.subplots()
ax.plot(bin_edges1, hist1, color='red', linewidth=2, label=".normal")
ax.plot(bin_edges2, hist2, color='green', linewidth=2, label=".standard_normal")
ax.set(xlabel='Data Values', ylabel='Data Counts', title='2 ways of defining non-standard normal distr.')
ax.legend(framealpha=1)
ax.grid()
plt.show()
# compare measures 
print(f"Ground truth: mean is {meanval}, std is {stdval}")
print(f".normal: mean is {np.mean(data1)}, std is {np.std(data1)}")
print(f".standard_normal: mean is {np.mean(data2)}, std is {np.std(data2)}")
```

```python
## overlay the mean

# compute the means
mean_d1 = np.mean(d1)
mean_d2 = np.mean(d2)
mean_d3 = np.mean(d3)

# plot them
plt.plot(x1, y1, 'b', x2, y2, 'r', x3, y3, 'k')
plt.plot([mean_d1, mean_d1], [0, np.max(y1)], 'b--')
plt.plot([mean_d2, mean_d2], [0, np.max(y2)], 'r--')
plt.plot([mean_d3, mean_d3], [0, np.max(y3)], 'k--')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.grid()
plt.show()
```

```python
## now for the standard deviation

# initialize
stds = np.zeros(3)

# compute standard deviations
stds[0] = np.std(d1, ddof=1)
stds[1] = np.std(d2, ddof=1)
stds[2] = np.std(d3, ddof=1)
#     numpy.std
# ddof means Delta Degrees of Freedom. The divisor used in calculations is N - ddof, 
# where N represents the number of elements. 
# By default ddof is zero <-> compute std for some population.

# same plot as earlier
plt.plot(x1, y1, 'b', x2, y2, 'r', x3, y3, 'k')
plt.plot([mean_d1, mean_d1], [0, np.max(y1)], 'b--', 
         [mean_d2, mean_d2], [0, np.max(y2)], 'r--',
         [mean_d3, mean_d3], [0, np.max(y3)], 'k--')
# now add stds
plt.plot([mean_d1-stds[0], mean_d1+stds[0]], [.4*np.max(y1), .4*np.max(y1)], 'b', linewidth=2)
plt.plot([mean_d2-stds[1], mean_d2+stds[1]], [.5*np.max(y2), .5*np.max(y2)], 'r', linewidth=2)
plt.plot([mean_d3-stds[2], mean_d3+stds[2]], [.6*np.max(y3), .6*np.max(y3)], 'k', linewidth=2)
# make it prettier
plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.grid()
plt.show()
```

```python
## Fano factor and coefficient of variation (CV)

# need positive-valued data 
# (because Fano and CV make sense only in that case: mu must be >0)
data = rng.poisson(3, 300)  # "Poisson noise"

fig, ax = plt.subplots(2, 1)
ax[0].plot(data, 's')
ax[0].set_title('Poisson noise')
ax[0].grid()

ax[1].hist(data)
ax[1].grid()
plt.show()
```

```python
## compute fano factor and coefficient of variation (CV) for a range of lambda parameters

# list of parameters
lambdas = np.linspace(1, 12, 15)

# initialize output vectors
fano = np.zeros(len(lambdas))
cv = np.zeros(len(lambdas))

for li in range(len(lambdas)):
    # generate new data
    data = rng.poisson(lambdas[li], 1000)

    # compute the metrics (need ddof=1 here?)
    fano[li] = np.var(data) / np.mean(data)
    cv[li] = np.std(data) / np.mean(data)  

# and plot
plt.plot(lambdas, cv, "bs-")
plt.plot(lambdas, fano, "ro-")
plt.legend(("CV", "Fano"))
plt.xlabel("$\lambda$")
plt.ylabel("amplitude")
plt.grid()
plt.show()

# Variance = mean!
```
