```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## "discrete" entropy

# generate data
N = 1000
rng = np.random.default_rng()
numbers = np.ceil(8*rng.uniform(size=N)**2)
numbers[numbers == 7] = 4

# plot it
plt.plot(numbers, 'o');
```

```python
## "discrete" entropy

# generate data
N = 1000
# rng = np.random.default_rng()
numbers = np.ceil(8*rng.uniform(size=N)**2)

# counts
u = np.unique(numbers)  
# probabilities
probs = np.zeros(len(u))  
for ui in range(len(u)):
    probs[ui] = np.sum(numbers == u[ui]) / N
    
# compute entropy (np.finfo(float).eps is for case probs = 0)
entropee = - np.sum(probs * np.log2(probs + np.finfo(float).eps))

# plot
plt.bar(u, probs, edgecolor='black')
plt.title(f'Entropy = {np.round(entropee, 3)}')
plt.xlabel('Data value')
plt.ylabel('Probability')
plt.show()
```

```python
## for random variables

# create Brownian noise
N = 1123
brownnoise = np.cumsum(np.sign(rng.standard_normal(N)))

fig, ax = plt.subplots(2, 1, figsize=(6, 10))
fig.subplots_adjust(hspace=0.3)

ax[0].plot(brownnoise, zorder=2)
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Data value')
ax[0].set_title('Brownian noise')
ax[0].grid(zorder=1)

ax[1].hist(brownnoise, 30, edgecolor='black', zorder=4)
ax[1].set_xlabel('Data value')
ax[1].set_ylabel('Counts')
ax[1].grid(zorder=3)
plt.show()
```

```python
### now compute entropy
# number of bins
nbins = 20

# bin the data and convert to probability
bin_counts, bin_edges = np.histogram(brownnoise, nbins)
probs = bin_counts / np.sum(bin_counts)

# compute entropy
entro = - np.sum(probs*np.log2(probs + np.finfo(float).eps))

print('Entropy = %g'%entro)
```
