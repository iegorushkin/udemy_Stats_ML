```python
# import libraries
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import numpy as np
```

```python
## create the data

n = 1000
thresh = 5 # threshold for cropping data
rng = np.random.default_rng()

data = np.exp(rng.standard_normal(n))

temp = data > thresh  # массив размера n с boolean-значениями 
print(temp[:n:10])  # как он выглядит?
print(np.sum(temp))  # считаем количество значений True

# Все значения, превышающие thresh, заменяются на  thresh + *что-то маленькое*
# даёт скопление данных в районе 5
data[data > thresh] = thresh + 0.1*rng.standard_normal(np.sum(data > thresh))
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

ax1.hist(data, bins='fd', linewidth=1, edgecolor='black')
ax1.set(title="Matplotlib's histogram", xlabel="Value", ylabel="Count")
ax1.grid()

sns.violinplot(data=data, cut=0, orient='h', ax=ax2)
ax2.set(title="Seaborn's violin plot", ylabel="Value")
ax2.grid()

plt.show()
```

```python
# show histogram
plt.hist(data, 30)
plt.title('Histogram')
plt.show()

# show violin plot
plt.violinplot(data)
plt.title('Violin')
plt.show()
```
